# Cleaning

```python
date = [] # None * length
for idx, row in news_df.iterrows():
  date.append(row["publishedAt"][:10])
```

```python
def cleaning(df, stop_words):
    #Drop na
    df = df.dropna()
    #Remove irrelevent days
    df = df[df['nextPosition'] != 0]
    #remove non alpha
    df.replace({'description': r'[^a-zA-Z ]+'}, {'description': ''}, regex=True, inplace=True)
    #lower case
    df['description'] = df['description'].str.lower()
    #Stop words
    df['description'] = df['description'].apply(lambda x:' '.join(x for x in x.split() if x not in stop_words))
    #lemmatization
    df['description'] = df['description'].apply(lambda x:' '.join([Word(x).lemmatize() for x in x.split()]))
    #Reset index
    df=df.reset_index(drop=True)
    return df
```

```python
df["Delta"] = df["Close"]-df["Open"]
df["Position"] = [-1 if i<0 else 1 for i in df["Delta"]]
df["Next_position"] = df["Position"].shift(-1)
```

```python
#1 if the stock price increases the next day, -1 if falls, 0 if markets are closed
position = [] # None * length
for idx, row in news_df.iterrows():
  if(row["date"] in df.index):
    #print(df.loc[row["date"]])
    position.append(df.loc[row["date"]]["Next_position"])
  else:
    position.append(0)
```

# Naive Bayes Classifier

```python
train_data = df2.sample(frac=0.7)
test_data = df2.drop(train_data.index)
```

```python
up_words = " ".join(train_data[train_data["nextPosition"] == 1]["description"]).split()
down_words = " ".join(train_data[train_data["nextPosition"] == -1]["description"]).split()
total_words = set(up_words + down_words)
```

```python
wordcloud.WordCloud(background_color="white").generate(" ".join(up_words)).to_file("thing.png")
```

```python
word_num_up=dict()
word_num_down=dict()
for word in total_words:
    if(up_words.count(word) - down_words.count(word)>0):
      word_num_up[word] = up_words.count(word) - down_words.count(word)
    elif(up_words.count(word) - down_words.count(word)<0):
      word_num_down[word] = down_words.count(word)-up_words.count(word)
```

```python
word_prob=dict()
num_up = len(train_data[train_data['nextPosition']==1])
num_down = len(train_data[train_data['nextPosition']==-1])
for word in total_words:
    word_prob[word] = {
        'prob_up' : (up_words.count(word) +0.5)/(num_up+1),
        'prob_down' : (down_words.count(word) +0.5 )/(num_down+1)
    }
#word_prob
```

```python
#P(s) is our expectation, we assume that this is 0.5
#P(¬S) or P(notS) is 1-0.5 because we dont have any prior knowledge
#The E is the expectation, which is the probability of these words appearing
def predict(sentence):
    PDE_list = []
    PUE_list = []
    sentence_arr = sentence.split()
    for word in sentence_arr:
      if(word in total_words):
        PEU = word_prob[word]["prob_up"]
        PED = word_prob[word]["prob_down"]
        PUE = (PEU*0.5)/(PEU*0.5 + PED*0.5)
        PDE = 1-PUE
        PUE_list.append(PUE)
        PDE_list.append(PDE)
    #We need to take the logs and sum as otherwise we would have to
    #multiply the probabilities which can cause problems as numbers
    #are very small
    log_PUE=sum(np.log10(PUE_list))
    log_PDE=sum(np.log10(PDE_list))

    if log_PUE>log_PDE:
      return 1.0
    else:
      return -1.0
```

```python
#can also use sklearn.metrics.accuracy_score for better runtime however this is
#easier to understand for demo
#Doesn't actually make a difference if we have true positive and true negative
#as long one day and short one day are the same position, see note on financial
#instruments
tp,tn,fp,fn=0,0,0,0


for index,row in test_data.iterrows():
  prediction = predict(row["description"])
  if(prediction==row["nextPosition"]):
    if(prediction==1.0):
      tp+=1
    else:
      tn+=1
  else:
    if(prediction==1.0):
      fp+=1
    else:
      fn+=1
accuracy = (tp+tn)/(tp+tn+fp+fn)
accuracy
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)#, random_state = 0)
```

```python
df
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
from transformers import TFDistilBertForSequenceClassification, TFTrainer, TFTrainingArguments

training_args = TFTrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
    eval_steps = 10,
)
```

```python
with training_args.strategy.scope():
    model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = TFTrainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)

trainer.train()
```

```python
trainer.evaluate(test_dataset)
```
